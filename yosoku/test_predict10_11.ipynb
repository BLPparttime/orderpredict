```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Excelデータの読み込み
file_path1 = 'data【10月】生産性結果記入用紙_Ver.4.3 (1).xlsx'  # Excelファイルのパスを指定
sheet_name = '実績昼'  # 読み込むシート名を指定

file_path2='data【11月】生産性結果記入用紙_Ver.4.3.xlsx'
train_data = pd.read_excel(file_path1, sheet_name=sheet_name)
test_data = pd.read_excel(file_path2, sheet_name=sheet_name)
train_data.head()
```

```python
test_data.head()
```

```python
# 行列を逆にする
train_data = train_data.T
test_data = test_data.T
# 1行目を列名にする
train_data.columns = train_data.iloc[1]
test_data.columns = test_data.iloc[1]
# 0,1行目を削除
train_data = train_data.drop(train_data.index[[0, 1]])
test_data = test_data.drop(test_data.index[[0, 1]])
train_data.head()
```

```python
test_data.head()
```

```python
# 欠損値を探す
train_data.isnull().sum()

```

```python
test_data.isnull().sum()
```

```python
# 0,1,4,11,12,13,14,15列目を削除
train_data = train_data.drop(train_data.columns[[0, 1, 4,11,12,13,14,15]], axis=1)
train_data.head()
```

```python
# 0,1,4,11,12,13,14,15列目を削除
test_data = test_data.drop(test_data.columns[[0, 1, 4,11,12,13,14,15]], axis=1)
test_data.head()
```

```python
# NaNを全て0に変換
train_data = train_data.fillna(0)
train_data.head()
```

```python
# NaNを全て0に変換
test_data = test_data.fillna(0)
test_data.head()
```

```python
# 0列目のデータを表示
train_data.iloc[:, 0]
train_data
```

```python
#0列目のデータを表示
test_data.iloc[:, 0]
test_data
```

```python
# 34行目を削除
train_data = train_data.drop(train_data.index[[31]])
train_data
```

```python
#34行目を削除
test_data = test_data.drop(test_data.index[[31]])
test_data
```

```python
 # 正規化(-1から1の範囲にする)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-1, 1))
train_data2 = scaler.fit_transform(train_data.iloc[:, 0:])
train_data2 = pd.DataFrame(train_data2, columns=train_data.columns[0:])
train_data2
```

```python
# 正規化(-1から1の範囲にする)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(-1, 1))
test_data2 = scaler.fit_transform(test_data.iloc[:, 0:])
test_data2 = pd.DataFrame(test_data2, columns=test_data.columns[0:])
test_data2
```

```python
# 目的変数と説明変数の設定
# 'target_column'を目的変数として、他のすべてのカラムを説明変数とする
X_train = train_data2.drop(columns= '件数')
y_train = train_data2['件数']
X_test = test_data2.drop(columns='件数')
y_test = test_data2['件数']

# ランダムフォレスト回帰モデルの構築
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 予測
y_pred = model.predict(X_test)

# モデルの評価
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R² Score: {r2}')
```

```python
#実際に予測
X_new = test_data2.drop(columns='件数')
y_new = model.predict(X_new)
print(f'予測結果: {y_new}')
#正規化を元に戻す
# 正規化前の値に戻すためのMinMaxScaler
y_scaler = MinMaxScaler(feature_range=(-1, 1))
y_scaler.fit(test_data[['件数']])

# 予測値を正規化前の値に戻す
y_new_original = y_scaler.inverse_transform([[y_new[0]]])

# 3行目の実際の値も正規化前に戻す
y_test_original = y_scaler.inverse_transform(y_test.values.reshape(-1, 1))

print(f'Predicted (元の値): {y_new_original[0][0]:.2f}')
print(f'Actual (元の値): {y_test_original[0][0]:.2f}')
print(f'Error (元の値): {y_test_original[0][0] - y_new_original[0][0]:.2f}')
```

```python
#特徴量の重要度を表示
importances = model.feature_importances_
feature_importances = pd.DataFrame(importances, index=X_train.columns, columns=['importance'])
feature_importances = feature_importances.sort_values(by='importance', ascending=False)
feature_importances
```

```python
# 特徴量重要度を可視化
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib_fontja
plt.figure(figsize=(12, 8))
sns.barplot(x='importance', y='features', data=feature_importances)
plt.title('特徴量重要度')
plt.show()
```
